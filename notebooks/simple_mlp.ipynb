## 4) Troubleshooting
- If the kernel cannot import `tensorflow` or `sklearn`, make sure you activated the same venv used to start Jupyter.
- On macOS, if you see errors installing TensorFlow, consider installing a CPU wheel that matches your macOS/Python or use conda (recommended for complex binary dependencies).
- For reproducibility, prefer using the `requirements.txt` file in the repo root.
