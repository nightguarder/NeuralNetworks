## 1) Recommended environment (terminal instructions)
Run these commands in a terminal (from the repository root). They create an isolated venv and install required packages.

```bash
# create and activate a venv (macOS / zsh)
python3 -m venv .venv
source .venv/bin/activate

# upgrade pip and install dependencies
python -m pip install --upgrade pip
python -m pip install -r requirements.txt

# start Jupyter Lab / Notebook (optional)
jupyter lab  # or `jupyter notebook`
```

## 4) Notes and troubleshooting
- If the kernel cannot import `tensorflow` or `sklearn`, make sure you activated the same venv used to start Jupyter.
- On macOS, if you see errors installing TensorFlow, consider installing a CPU wheel that matches your macOS/Python or use conda (recommended for complex binary dependencies).
- For reproducibility, prefer using the `requirements.txt` file in the repo root.
